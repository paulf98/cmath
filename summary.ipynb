# Parameter Curves

## 2-D Parameter Curves

A 2-D Parameter Curve is defined as $f(t) = x, g(t) = y,  t ∈ [a, b]$

A resulting image set is called a trace of a parameter curve.

As an example, consider the following parameter curve:

$f(t) = cos(t)$

$g(t) = sin(t)$

$t ∈  [0, 2\pi] $

```python
import numpy as np
import matplotlib.pyplot as plt

t = np.arange(0, 2*np.pi, 0.001)
x = np.cos(t)
y = np.sin(t)

plt.axis('equal')
plt.grid()
plt.plot(x, y)
```

The trace of the Parameter Curve is the unit circle. If we change the Parameter Curve to:

$f(t) = cos(2\pi t)$

$g(t) = sin(2\pi t)$

$t ∈  [0, 1] $

The trace of the Parameter Curve is the unit circle again. So we see, the trace of a parameter curve is independent of the parameterization.

```python
t = np.arange(0, 1, 0.001)
x = np.cos(2*np.pi*t)
y = np.sin(2*np.pi*t)

plt.axis('equal')
plt.grid()
plt.plot(x, y)
```

## Velocity and Acceleration

The interesting thing now is the speed at which the circle is traversed.

The **velocity** is defined as:

$K'(t): [a, b] \rightarrow \mathbb{R}^2, t \mapsto \begin{pmatrix} f'(t) \\ g'(t) \end{pmatrix}$

The **web speed** is defined as:

$v(t) = || K'(t) ||$

The **acceleration** is defined as:

$K''(t): [a, b] \rightarrow \mathbb{R}^2, t \mapsto \begin{pmatrix} f''(t) \\ g''(t) \end{pmatrix}$

```python
# velocity
t = np.arange(0, 1, 0.001)
x = -2 * np.pi * np.sin( 2 * np.pi * t )
y = 2 * np.pi * np.cos( 2 * np.pi * t )

plt.axis('equal')
plt.grid()
plt.title('Velocity')
plt.plot(x, y)
```

```python
# web speed
t = np.arange(0, 1, 0.001)
x = -2 * np.pi * np.sin( 2 * np.pi * t )
y = 2 * np.pi * np.cos( 2 * np.pi * t )

v = np.linalg.norm([x, y], axis=0)
  
plt.grid()
plt.title('Web Speed')
plt.plot(v)
```

The web speed is constant at $2\pi$.

```python
# acceleration
t = np.arange(0, 1, 0.001)
x = -4 * np.power(np.pi, 2) * np.cos( 2 * np.pi * t )
y = -4 * np.power(np.pi, 2) * np.sin( 2 * np.pi * t )

plt.axis('equal')
plt.grid()
plt.title('Acceleration')
plt.plot(x, y)
```

As you can see, the graphs of velocity and acceleration are not very exciting. In the following, we will look at another example.

### Semicubical parabola

$f(t) = t^2$

$g(t) = t^3$

$t ∈  \mathbb{R}$

```python
t = np.linspace(-2, 2, 100)
x = np.power(t, 2)
y = np.power(t, 3)


plt.axis('equal')
# show x an y axis
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')

plt.grid()
plt.plot(x, y)
```

### Velocity and Acceleration

$f'(t) = 2t$

$g'(t) = 3t^2$

```python
ft1 =  2 * t
gt1 =  3 * np.power(t, 2)

vt1 = np.linalg.norm([ft1, gt1], axis=0)

plt.axis('equal')
# show x an y axis
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')

plt.grid()
plt.plot(t, vt1)
```

The Semi-cubical parabola has a special point in the graph of the velocity, because it has a 0-value. This means the curve can't be regular.

## Curves with web speed=1

Another special type of curves are those with a constant velocity of 1. These are called **unit speed curves**. An example is the following:

$f(t) = r * cos(\frac{t}{r})$

$g(t) = r * sin(\frac{t}{r})$

$t ∈  [0, \mathbb{2\pi r}]$

```python
r = 1
t = np.linspace(0, 2 * np.pi * r, 100)
ft1 = - r * np.sin(t / r)
gt1 = r * np.cos(t / r)

vt = np.linalg.norm([ft1, gt1], axis=0)

plt.axis('equal')
# show x an y axis
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')

plt.grid()
plt.plot(t, vt)
```

## Tangent and Normal

Now we know about velocity and acceleration, we can define the **tangent** and **normal** to a curve.

The **tangent** or **tangent vector** is defined as:

$\vec T(t) = \frac{1}{v(t)} \begin{pmatrix} f'(t) \\ g'(t) \end{pmatrix}$

We can also define the **normal** or **normal vector** as:

$\vec N(t) = \frac{1}{v(t)} \begin{pmatrix} -g'(t) \\ f'(t) \end{pmatrix}$

The **tangent** and **normal** are orthogonal to each other. Note that the normal vector rotates the tangent vector left by $90^{\circ}$. You can use the right hand rule to verify this.

Here's an example of the unit circle with some tangents and normals:

```python
t = np.linspace( 0 , 2 * np.pi , 200 )  
x = np.cos( t ) 
y = np.sin( t ) 
 
figure, axes = plt.subplots( 1 ) 
 
axes.plot( x, y ) 
axes.set_aspect( 1 ) 

plt.arrow(1.0, 0.0, 0.0, 1.0, color="g", head_width=0.05)
plt.arrow(1.0, 0.0, -0.9, 0.0, color="r", head_width=0.05)

plt.arrow(0.0, 1.0, -1.0, 0.0, color="g", head_width=0.05)
plt.arrow(0.0, 1.0, 0.0, -0.9, color="r", head_width=0.05)

plt.arrow(-1.0, 0.0, 0.0, -1.0, color="g", head_width=0.05)
plt.arrow(-1.0, 0.0, 0.9, 0.0, color="r", head_width=0.05)

plt.arrow(0.0, -1.0, 1.0, 0.0, color="g", head_width=0.05)
plt.arrow(0.0, -1.0, 0.0, 0.9, color="r", head_width=0.05)

arrowxy = 0.5*np.sqrt(2.0)
plt.arrow(arrowxy, arrowxy,-arrowxy, arrowxy,color="g",head_width=0.05)

plt.title( 'Parametric Equation Circle' ) 
plt.show() 
```
