# Parameter Curves

## 2-D Parameter Curves

A 2-D Parameter Curve is defined as $f(t) = x, g(t) = y,  t ∈ [a, b]$

A resulting image set is called a trace of a parameter curve.

As an example, consider the following parameter curve:

$f(t) = cos(t)$

$g(t) = sin(t)$

$t ∈  [0, 2\pi] $

```python
import numpy as np
import matplotlib.pyplot as plt

t = np.arange(0, 2*np.pi, 0.001)
x = np.cos(t)
y = np.sin(t)

plt.axis('equal')
plt.grid()
plt.plot(x, y)
```

The trace of the Parameter Curve is the unit circle. If we change the Parameter Curve to:

$f(t) = cos(2\pi t)$

$g(t) = sin(2\pi t)$

$t ∈  [0, 1] $

The trace of the Parameter Curve is the unit circle again. So we see, the trace of a parameter curve is independent of the parameterization.

```python
t = np.arange(0, 1, 0.001)
x = np.cos(2*np.pi*t)
y = np.sin(2*np.pi*t)

plt.axis('equal')
plt.grid()
plt.plot(x, y)
```
