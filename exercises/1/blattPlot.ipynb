```python
import matplotlib.pyplot as plt 
import numpy as np

x = np.arange( 0 , 2 * np.pi, .1 )
y = np.sin( x )

figure, axes = plt.subplots( 1 ) 
 
axes.plot( x, y ) 
axes.set_aspect( 1 ) 

plt.title ( "Sinuskurve" )
plt.show()
```

```python
def draw_fx (x, y, title = ""):
    figure, axes = plt.subplots( 1 ) 
    axes.plot( x, y ) 

    plt.title ( title )
    plt.show()

x = np.arange(-3, 3, .1)

y = np.power(x, 2)
draw_fx(x, y, "x^2")

y = np.power(x, 3)
draw_fx(x, y, "x^3")

y = np.sin(x)
draw_fx(x, y, "sin(x)")

y = np.power(np.e, x)
draw_fx(x, y, "e^x")
```

```python
# Aufgabe 1)d)

y = np.power(x, 3) - (2 * np.power(x, 2)) + 3

draw_fx(x, y, "Polynom")
```

## Plotly

```python
import plotly.express as px

def draw_fx (x, y, title = ""):
    fig = px.line(x=x, y=y, title=title)
    fig.show()

x = np.arange(-3, 3, .1)

y = np.power(x, 2)
draw_fx(x, y, "x^2")

y = np.power(x, 3)
draw_fx(x, y, "x^3")

y = np.sin(x)
draw_fx(x, y, "sin(x)")

y = np.power(np.e, x)
draw_fx(x, y, "e^x")
```

```python
x = np.arange(-4, 4, .1)
y = (1 / np.sqrt(2 * np.pi)) * (np.power(np.e, - (np.divide(1, 2) * np.power(x, 2))))

draw_fx(x, y)
```
