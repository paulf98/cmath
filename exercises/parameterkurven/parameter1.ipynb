```python
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(-5, 5, 100)
ft = np.power(t, 3) - 2 * t
gt = np.power(t, 2) - t


# Horizontal and vertical tangents
plt.arrow(0.0, 1.8, 0, 5, color="g", head_width=0.5)
plt.arrow(0.0, -3.8, 0, 5, color="g", head_width=0.5)

plt.arrow(-1.25, 0, 5, 0, color="g", head_width=0.5)

plt.plot(ft, gt)
```

```python
ft1 = 3 * np.power(t, 2) - 2
gt1 = 2  * t - 1

# Show the x and y axis as lines
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')

plt.plot(ft1, gt1)
```
