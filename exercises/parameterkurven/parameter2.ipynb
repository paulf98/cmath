```python
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(0, 2*np.pi, 100)

def plot_a():
    ft = 3 * np.power(np.cos(t), 3)
    gt = 3 * np.power(np.sin(t), 3)
    plt.plot(ft, gt)

def plot_b():
    t = np.linspace(0, 6*np.pi, 100)
    ft = 0.8 * np.power(np.e, 0.05 * t) * np.cos(t)
    gt = 0.8 * np.power(np.e, 0.05 * t) * np.sin(t)
    plt.plot(ft, gt)
    
def plot_c():
    ft = np.sin(t)
    gt = np.power(np.sin(t), 2)
    plt.plot(ft, gt)

def plot_d():
    t = np.linspace(-1.7, 1.7, 100)
    ft = np.power(t, 2)
    gt = np.power(t, 3) - 2 * t
    plt.plot(ft, gt)

plot_a() 
# plot_b()
# plot_c()
# plot_d()
```

## Aufgabe 4)

```python
def plot_a():
    t = np.linspace(-2*np.pi, 2*np.pi, 100)
    ft = np.power(t, 3) - 2 * t
    gt = np.power(t, 2) - t
    plt.plot(ft, gt)


plot_a()
```

```python
def plot_b():
    t = np.linspace(0.1, 2, 100)
    ft = np.power(t, 4) - np.power(t, 2)
    gt = t + np.log10(t)
    plt.plot(ft, gt)

plot_b()
```

```python
def plot_c():
    t = np.linspace(-2*np.pi, 2*np.pi, 100)
    ft = np.sin(3 * t)
    gt = np.sin(4 * t)
    plt.plot(ft, gt)

plot_c()
```

```python
def plot_d():
    t = np.linspace(-2*np.pi, 2*np.pi, 100)
    ft = t + np.sin(2 * t)
    gt = t + np.sin(3 * t)
    plt.plot(ft, gt)

plot_d()
```

```python
def plot_e():
    t = np.linspace(-2*np.pi, 2*np.pi, 100)
    ft = np.sin(t + np.sin(t))
    gt = np.cos(t + np.cos(t))
    plt.plot(ft, gt)

plot_e()
```

```python
def plot_f():
    t = np.linspace(-2*np.pi, 2*np.pi, 1000)
    ft = np.cos(t)
    gt = np.sin(t + np.sin(5 * t))
    plt.plot(ft, gt)

plot_f()
```
