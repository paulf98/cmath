# Kurven in Polarkoordinaten

## Aufgabe 1)


```python
import numpy as np
import matplotlib.pyplot as plt

phi = np.linspace(-np.pi, np.pi, 100)

# a)
r = 1 + np.cos(phi)
# plt.polar(phi, r, '-g')

# b)
r = 1 - np.cos(phi)
# plt.polar(phi, r, '-g')

# c)
r = 1 - np.sin(phi)
plt.polar(phi, r, '-g')
```
