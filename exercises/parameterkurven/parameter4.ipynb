## Aufgabe 1)

```python
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(-1.4, 1.4, 100)
ft = np.power(t, 2)
gt = np.power(t, 3) - 3 * t

plt.axis('equal')
plt.axvline(0, color='k')
plt.legend()
plt.plot(ft, gt, )
```

# Reparametrisierung mit umgedrehter Durchlaufrichtung

```python
t = np.linspace(1.4, -1.4, 100)
ft = np.power(-t, 2)
gt = np.power(-t, 3) + 3 * t

plt.axis('equal')
plt.axvline(0, color='k')
plt.legend()
plt.plot(ft, gt, )
```

## Aufgabe 2)

```python
t = np.linspace(-2, 2, 100)
ft = np.power(t, 4) - 1
gt = np.power(t, 3) + 1

plt.axis('equal')
plt.axvline(0, color='k')
plt.axhline(0, color='k')
plt.plot(ft, gt,)
```

```python
ft1 = 4 * np.power(t, 3)
gt1 = 3 * np.power(t, 2)

vt = np.sqrt(np.power(ft1, 2) + np.power(gt1, 2))
plt.axhline(0, color='k')
plt.axvline(0, color='k')
plt.plot(t, vt, label='v(t)')
```
