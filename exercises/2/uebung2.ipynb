# Aufgabe 1

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-6, 6, 100)
y = np.sin(x)

plt.plot(x, y)
plt.show()
```

## b)

```python
def fx(x, y):
    plt.plot(x, y)
    plt.show()

fx(x, y)
```

## c)

```python
y = np.power(x, 2)
fx(x, y)

y = np.power(x, 3)
fx(x, y)

y = np.sin(x)
fx(x, y)

y = np.power(np.e, x)
fx(x, y)
```

## d)

```python
y = np.power(x, 3) - 2 * np.power(x, 2) + 3
fx(x, y)
```

```python
import plotly.express as px

y = np.sin(x)
px.line(x=x, y=y).show()
```

## b) 

```python
def fx(x, y):
    px.line(x=x, y=y).show()
```

## c)

```python
y = np.power(x, 2)
fx(x, y)

y = np.power(x, 3)
fx(x, y)

y = np.power(np.e, x)
fx(x, y)
```

## d)

```python
y = np.power(x, 3) - 2 * np.power(x, 2) + 3
fx(x, y)
```

# Aufgabe 3)

```python
x = np.linspace(-4, 4, 100)
y = 1 / np.sqrt(2 * np.pi) * np.power(np.e, (-1/2 * np.power(x, 2)))

fx(x, y)
```
