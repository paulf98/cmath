### a) 

für x < 0 ist der Abstand zur x-Achse. Für x >= 0 ist der Wert der Abstand zum Ursprung. Fallunterscheidung zwischen erster/vierter Quadrant und zweiter/dritter Quadrant.

### b) 

Negative Konturlinien sind nicht sinnvoll. Der Abstand zum Ursprung ist immer positiv.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-2, 2, 100)
y = np.linspace(-2, 2, 100)
x, y = np.meshgrid(x, y)

# draw contour lines for the function f(x,y) = sqrt(x^2 + y^2) if x >= 0, else f(x, y) = |y|
z = np.where(x >= 0.0, np.sqrt(x*x + y*y), np.abs(y))

# plot the contour lines at the values 0, 1 and 2
plt.contourf(x, y, z, 100, cmap='jet')
plt.contour(x, y, z, [0, 1, 2])

fig = plt.figure(figsize=(16.0, 9.0))
ax = fig.add_subplot(111, projection='3d')

lw = 0.8
ax.plot_surface(x, y, z, rstride=1, cstride=1, cmap='jet', linewidth=lw, antialiased=True, alpha=0.75)

ax.set_title('f(x,y) = sqrt(x^2 + y^2) if x >= 0, else f(x, y) = |y|', y=1.1, fontsize=24)
plt.xlabel('x')
plt.ylabel('y')
ax.set_zlabel('z')
```
