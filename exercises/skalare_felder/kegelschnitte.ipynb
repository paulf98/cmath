# Linearformen

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import linalg

x = np.linspace(-10, 10, 100)
y = np.linspace(-10, 10, 100)

X, Y = np.meshgrid(x, y)
Z = X + Y

plt.axes().set_aspect('equal')
plt.contour(X, Y, Z, 10)

fig = plt.figure(figsize=(10.0, 10.0))
ax = fig.add_subplot(111, projection='3d')

lw = 0.8
ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap='jet', linewidth=lw, antialiased=True, alpha=0.75)

plt.xlabel('x')
plt.ylabel('y')
ax.set_zlabel('z')
```

# Quadratische Formen

```python
# Ellipse - Zwei Eigenwerte mit gleichem Vorzeichen
Z = np.power(X, 2) + 2 * np.power(Y, 2)

plt.contour(X, Y, Z, 10)
```

```python
# Parabel - Nur ein Eigenwert != 0
Z = X - 2 * np.power(Y, 2)

plt.contour(X, Y, Z, 10)
```

## Kegelschnitte

```python
# Kegelschnitte
a = np.array([[5.0, -3.0], [-3.0, 5.0]])
b = np.array([-12.0, -12.0])
c, low = linalg.cho_factor(a)
x = linalg.cho_solve((c, low), -0.5*b)

print('Der Verschiebungsfaktor der quadratischen Form ist: ')
print(x)
```

## Beispiel

Eine quadratische Form:

```python
def quadric(x, y, a=1.0, b=0.0, c=1.0, d=np.array([0, 0]), f=0.0):
    """ Quadratische Form """
    return a*x*x + 2.0*b*x*y + c*y*y + d[0]*x + d[1]*y
```

```python
x = np.linspace(-10, 10, 100)
y = np.linspace(-10, 10, 100)

X, Y = np.meshgrid(x, y)
# Z = quadric(X, Y, b=-0.5)
Z = quadric(X, Y, a= 5.0, b=-3, c=5.0, d=[-12, -12], f=-8.0)

plt.contour(X, Y, Z, 10)

fig = plt.figure(figsize=(10.0, 10.0))
ax = fig.add_subplot(111, projection='3d')

lw = 0.8
ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap='jet', linewidth=lw, antialiased=True, alpha=0.75)

plt.xlabel('x')
plt.ylabel('y')
ax.set_zlabel('z')
```
