# Skalare Felder

```python
import matplotlib.pyplot as plt
import numpy as np

# Create a 2D array of points
X, Y = np.meshgrid(np.arange(-5, 5, 0.01), np.arange(-5, 5, 0.01))
Z = np.fmax(np.abs(X), np.abs(Y))

# Draw the contour lines
plt.contour(X, Y, Z, levels=np.arange(0, 2.5, 0.5))

# Show the plot
plt.grid()
plt.show()
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 50, cmap='plasma')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
```

```python
# Create a 2D array of points
X, Y = np.meshgrid(np.arange(0, 10, 0.01), np.arange(0, 10, 0.01))
Z = np.power(np.e, X) * np.sin(Y)

# Draw the contour lines
plt.contour(X, Y, Z, levels=np.arange(0, 2.5, 0.5))

# Show the plot
plt.grid()
plt.show()
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 50, cmap='plasma')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');
```
