## Aufgabe 1 - Schiefer Wurf

```python
import numpy as np
import matplotlib.pyplot as plt

g = 9.81

def crooked_throw(v0, alpha, t):
    ft = v0 * np.cos(alpha) * t
    gt = v0 * np.sin(alpha) * t - 0.5 * g * t**2
    plt.plot(ft, gt)

v0 = 500
# convert to radians
alpha = 30 * np.pi / 180.
t = np.linspace(0, 100, 100)

plt.ylim(0, 10000)
plt.xlim(0, 25000)
crooked_throw(v0, alpha, t)
```
