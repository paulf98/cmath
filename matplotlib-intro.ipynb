## Fortgeschrittene Plots

Hier wird LaTex-Code eingebunden.

```python
import numpy as np
x = np.arange(1.0, 19.0, 0.4)
y1 = np.log10(x)
y2 = 0.01*x**2
y3 = 0.9*np.sin(x)
plt.plot(x, y1, 'r-', label='$y_1=log_{10}(x)$')
plt.plot(x, y2, 'b.', label='$y_2=0.01 \cdot x^2$')
plt.plot(x, y3, 'go', label='$y_3=0.9 \cdot sin(x)$')
plt.plot(x, y1+y2+y3, '+', label='$y_4 = y_1+y_2+y_3$')
plt.axis([0, 20, -1, 5])
plt.title('Mehrere Funktionen in einer Grafik')
plt.legend(title='Funktionen', loc='upper left')
```

# Histogramme

```python
np.random.seed(42)
x = np.random.randn(500)

histmin = np.floor(min(x))
histmax = np.ceil(max(x))
width = 0.5
# Klassen bilden
classes = np.arange(histmin, histmax, width)
plt.hist(x, bins=classes)
plt.show()
```

## Skalare Felder

```python
n = 200
x = np.linspace(0, 1.0*np.pi, n)
y = np.linspace(0, 2.0*np.pi, n)
x, y = np.meshgrid(x, y)
z = np.sin(x)*np.sin(y)

fig = plt.figure()
ax = plt.imshow(z, cmap='jet', origin='lower')
plt.title('Eine Heatmap')
plt.xlabel('x')
plt.ylabel('y')
plt.axis([0, 200, 0, 200])
plt.xticks(np.arange(0, 201, 25))
plt.yticks(np.arange(0, 201, 25))

# Eine Legende mit Hilfe von colorbar
cbar = fig.colorbar(ax)
cbar.ax.get_yaxis().labelpad = 15
cbar.ax.set_ylabel('Funktionswerte', rotation=270)
```

## Mehrere Grafiken mit Subplots

```python
def func(x):
    """A simple function"""
    return np.sin(2.0*np.pi*x)

x1 = np.arange(0.0, 4.0, 0.1)
x2 = np.arange(0.0, 4.0, 0.01)
y1 = func(x1)
y2 = func(x2)
y1n = y1 + 0.1*np.random.randn(len(x1))

fig = plt.figure()
plt.subplot(211)
plt.plot(x1, y1, 'bo', x2, y2, 'r:')
plt.subplot(212)
plt.plot(x1, y1n, 'bo', x2, y2, 'r:')
```

## 3-Dimensionale Plots
